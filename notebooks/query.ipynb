```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('../data/processed/kavak_market.db')

query = """
SELECT marca, COUNT(*) AS total, AVG(precio_mxn) as precio_promedio 
FROM autos
GROUP BY marca
ORDER BY total DESC;
"""

df = pd.read_sql_query(query, conn)
conn.close()
```
